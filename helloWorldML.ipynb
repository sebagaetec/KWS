```python
import sys
import numpy as np
import tensorflow as tf
```

Para compatibilidad

```python
if tf.__version__.split('.')[0] != '2':
    raise Exception((f"The script is developed and tested for tensorflow 2. "
                     f"Current version: {tf.__version__}"))

if sys.version_info.major < 3:
    raise Exception((f"The script is developed and tested for Python 3. "
                     f"Current version: {sys.version_info.major}"))
```

Modelo

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])

model.compile(optimizer='sgd', loss='mean_squared_error')

#input
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
#output
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

#ajustar el modelo a los datos (entrenamiento)
model.fit(xs, ys, epochs=250)
```

Predicción

```python
#relacion "real" entre xs ys es f(x)= 2x - 1, para 10 la salida deberia ser 19
print(model.predict(np.array([10.0])))
```

Convertir modelo a .tflite

```python
export_dir = 'saved_model/1'
tf.saved_model.save(model, export_dir)
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model(export_dir)
tflite_model = converter.convert()
```

Tamaño modelo Original

```python
from pathlib import Path
model_path = Path("saved_model/1/saved_model.pb")
size_bytes = model_path.stat().st_size
print(f"Tamaño de saved_model.pb: {size_bytes} bytes")
```

Tamaña modelo convertido

```python
import pathlib
tflite_model_file = pathlib.Path('model.tflite')
tflite_model_file.write_bytes(tflite_model)
```
